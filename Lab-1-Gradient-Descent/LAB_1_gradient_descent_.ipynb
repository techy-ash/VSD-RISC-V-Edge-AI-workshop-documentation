```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("studentscores.csv")
plt.scatter(
    dataset["Hours"],
    dataset["Scores"],
    color = "red"
)
plt.show()
```

```python
X = dataset.iloc[:, :1].values
Y = dataset.iloc[:, 1].values
```

```python
print(X)
```

y = mx + c #x is a vector m and c are constants

```python
class Model():
    def __init__(self, learning_rate, iterations):
        self.learning_rate = learning_rate
        self.iterations = iterations

    def predict(self, X):
      # calculating y = mx + c
      return X.dot(self.slope) + self.const

    def fit(self, X, Y):
      self.m, self.n = X.shape
      self.slope = np.zeros(self.n)
      self.const = 0
      self.X = X
      self.Y = Y

      for i in range(self.iterations):
        self.update_weights()
      return self

    def update_weights(self):
      Y_pred = self.predict(self.X)
      dW = - (2 * (self.X.T).dot(self.Y - Y_pred)) / self.m
      db = - 2 * np.sum(self.Y - Y_pred) / self.m

      self.slope -= self.learning_rate * dW
      self.const -= self.learning_rate * db
      return self
```

```python
model = Model(learning_rate = 0.01, iterations = 1000)
model.fit(X,Y)
```

```python
Y_pred = model.predict(X)
print(Y_pred)
```

```python
plt.scatter(dataset["Hours"], dataset["Scores"], color = "blue")
plt.scatter(X, Y_pred, color = "red")
plt.show()
```

```python
print(model.slope, model.const)
```

```python
dataset = pd.read_csv("50_Startups.csv")
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
print(x)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

regressor = LinearRegression()
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
print(y_pred)
```

```python
for i, (pred, actual) in enumerate(zip(y_pred, y_test)):
  print(f"Sample {i+1}: Predicted = {pred:.2f}, Actual = {actual:.2f}")
```

```python
print("Coefficients:", regressor.coef_)
print("Intercept:", regressor.intercept_)
```

```python
#polynomial regression
dataset = pd.read_csv("studentscores.csv")
plt.scatter(dataset["Hours"], dataset["Scores"])
plt.show()
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree = 3)
x_poly = poly.fit_transform(x)
print(x_poly)
```

```python
model = LinearRegression()
model.fit(x_poly, y)
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
y_pred = model.predict(x_poly)
print(y_pred)
```
